```python
# Load the COCO JSON file
coco_json_file = 'path/to/your/coco.json'
with open(coco_json_file, 'r') as f:
    coco_data = json.load(f)

# YOLO class names (replace with your own)
class_names = ['class1', 'class2', 'class3']

# Function to convert COCO bounding box to YOLO format
def coco_to_yolo_bbox(coco_bbox, image_width, image_height):
    x, y, w, h = coco_bbox
    x_center = x + (w / 2)
    y_center = y + (h / 2)
    x_center /= image_width
    y_center /= image_height
    w /= image_width
    h /= image_height
    return x_center, y_center, w, h

# Loop through each image annotation
for image_info in coco_data['images']:
    image_id = image_info['id']
    image_width = image_info['width']
    image_height = image_info['height']

    # Create YOLO TXT file for this image
    yolo_txt_filename = os.path.splitext(image_info['file_name'])[0] + '.txt'
    with open(yolo_txt_filename, 'w') as txt_file:
        for annotation in coco_data['annotations']:
            if annotation['image_id'] == image_id:
                category_id = annotation['category_id']
                class_id = category_id - 1  # COCO class IDs start from 1

                # Convert COCO bounding box to YOLO format
                x_center, y_center, w, h = coco_to_yolo_bbox(
                    annotation['bbox'], image_width, image_height)

                # Write YOLO annotation to the TXT file
                txt_file.write(f"{class_id} {x_center} {y_center} {w} {h}\n")

print("Conversion complete.")
```
